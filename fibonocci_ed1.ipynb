```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Expanding the code to generalize n_f and n_t

nf = 20
nt = 20

for ii in range(2,nf+1):

    n_f = ii #no. of terms to sum for fibonocci
    n_t = nt #total no. of fibonocci terms to be computed

    f_s = [] # list def for fibonocci series
    f_d = []
#first fibonocci term of the series, different from regular terms

    f_n = np.linspace(0,n_f,n_f, endpoint=False)
    f_s.append(sum(f_n))

##### find remaining terms of the series

    for i in range(2,n_t+1):
        if i<=n_f:
            f_add=sum(f_n[i-1:n_f+1]) + sum(f_s[0:i])
            f_s.append(f_add)
        else:
            f_s.append(sum(f_s[i-n_f-1:i-1]))

    for kk in range(1,nt-1):
        fd_add = f_s[kk+1]/f_s[kk]
        f_d.append(fd_add)

    n_s = np.linspace(0,nt,nt,endpoint = False)
    ns = n_s[ii:nt]

  #  print (f_d)
  #  print (ns)
  #  print (f_s)
    plt.subplot(121)
    plt.plot(n_s,f_s)
    plt.subplot(122)
    plt.plot(n_s[:nt-2],f_d)

#plt.hold(True)
plt.grid(True)
plt.show()
```
