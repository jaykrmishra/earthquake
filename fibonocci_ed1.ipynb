```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#warnings.filterwarnings(action='once')
##(to see warnings once)

#Expanding the code to generalize n_f and n_t

nf = 10 # beginning from 2
nt = 23

plt.figure(figsize=(20,20)) 

#fig, ax = plt.subplots(figsize=(20, 10)) 
#fig = plt.gcf() --> fig.set_size_inches(11,8)

f_r = []
f_nn = []
for ii in range(2,nf+1):

    n_f = ii #no. of terms to sum for fibonocci
    n_t = nt #total no. of fibonocci terms to be computed

    f_s = [] # list def for fibonocci series
    f_d = []
  
#first fibonocci term of the series, different from regular terms

    f_n = np.linspace(0,n_f,n_f, endpoint=False)
    f_s.append(sum(f_n))

##### find remaining terms of the series

    for i in range(2,n_t+1):
        if i<=n_f:
            f_add=sum(f_n[i-1:n_f+1]) + sum(f_s[0:i])
            f_s.append(f_add)
        else:
            f_s.append(sum(f_s[i-n_f-1:i-1]))

    for kk in range(1,nt-1):
        fd_add = f_s[kk+1]/f_s[kk]
        f_d.append(fd_add)

    n_s = np.linspace(0,nt,nt,endpoint = False)
    ns = n_s[ii:nt]

  #  print (f_d)
  #  print (ns)
  #  print (f_s)
    plt.subplot(221)
    plt.plot(n_s,f_s)
    plt.subplot(222)
    plt.plot(n_s[:nt-2],f_d)
    
       
    plt.subplot(221)
    plt.xlabel('nth fibonocci term')
    plt.ylabel('sum')
    plt.title('sum of fibonocci terms')
    plt.grid(True)
    #plt.legend(handles=n_f)
    
    plt.subplot(222)
    plt.xlabel('nth fibonocci term')
    plt.ylabel('fibonocci ratio')
    plt.title('fibonocci ratio as fn of order')
    plt.grid(True)
    
    f_r.append(f_d[nt-4]) # constant fibonocci ratio
    f_nn.append(ii)

plt.subplot(223)
plt.plot(f_nn,f_r)
plt.xlabel('fibonocci order')
plt.ylabel('fibonocci ratio')
plt.title('fibonocci order vs fibonocci ratio')
plt.grid(True)
    #    print (f_d)
#print (f_n+2, )

#plt.hold(True)
#plt.show()
```
